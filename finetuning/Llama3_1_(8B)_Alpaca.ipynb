### Unsloth

```python
from unsloth import FastLanguageModel
import torch
max_seq_length = 2048 # Choose any! We auto support RoPE Scaling internally!
dtype = None # None for auto detection. Float16 for Tesla T4, V100, Bfloat16 for Ampere+
load_in_4bit = True # Use 4bit quantization to reduce memory usage. Can be False.

# 4bit pre quantized models we support for 4x faster downloading + no OOMs.
fourbit_models = [
    "unsloth/Meta-Llama-3.1-8B-bnb-4bit",      # Llama-3.1 15 trillion tokens model 2x faster!
    "unsloth/Meta-Llama-3.1-8B-Instruct-bnb-4bit",
    "unsloth/Meta-Llama-3.1-70B-bnb-4bit",
    "unsloth/Meta-Llama-3.1-405B-bnb-4bit",    # We also uploaded 4bit for 405b!
    "unsloth/Mistral-Nemo-Base-2407-bnb-4bit", # New Mistral 12b 2x faster!
    "unsloth/Mistral-Nemo-Instruct-2407-bnb-4bit",
    "unsloth/mistral-7b-v0.3-bnb-4bit",        # Mistral v3 2x faster!
    "unsloth/mistral-7b-instruct-v0.3-bnb-4bit",
    "unsloth/Phi-3.5-mini-instruct",           # Phi-3.5 2x faster!
    "unsloth/Phi-3-medium-4k-instruct",
    "unsloth/gemma-2-9b-bnb-4bit",
    "unsloth/gemma-2-27b-bnb-4bit",            # Gemma 2x faster!
] # More models at https://huggingface.co/unsloth

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Meta-Llama-3.1-8B",
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
    # token = "hf_...", # use one if using gated models like meta-llama/Llama-2-7b-hf
)
```

We now add LoRA adapters so we only need to update 1 to 10% of all parameters!

```python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16, # Choose any number > 0 ! Suggested 8, 16, 32, 64, 128
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 16,
    lora_dropout = 0, # Supports any, but = 0 is optimized
    bias = "none",    # Supports any, but = "none" is optimized
    # [NEW] "unsloth" uses 30% less VRAM, fits 2x larger batch sizes!
    use_gradient_checkpointing = "unsloth", # True or "unsloth" for very long context
    random_state = 3407,
    use_rslora = False,  # We support rank stabilized LoRA
    loftq_config = None, # And LoftQ
)
```

<a name="Data"></a>
### Data Prep
We now use the Alpaca dataset from [yahma](https://huggingface.co/datasets/yahma/alpaca-cleaned), which is a filtered version of 52K of the original [Alpaca dataset](https://crfm.stanford.edu/2023/03/13/alpaca.html). You can replace this code section with your own data prep.

**[NOTE]** To train only on completions (ignoring the user's input) read TRL's docs [here](https://huggingface.co/docs/trl/sft_trainer#train-on-completions-only).

**[NOTE]** Remember to add the **EOS_TOKEN** to the tokenized output!! Otherwise you'll get infinite generations!

If you want to use the `llama-3` template for ShareGPT datasets, try our conversational [notebook](https://colab.research.google.com/github/unslothai/notebooks/blob/main/nb/Llama3_(8B)-Alpaca.ipynb)

For text completions like novel writing, try this [notebook](https://colab.research.google.com/github/unslothai/notebooks/blob/main/nb/Mistral_(7B)-Text_Completion.ipynb).

```python
alpaca_prompt = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.

### Instruction:
{}

### Input:
{}

### Response:
{}"""

EOS_TOKEN = tokenizer.eos_token # Must add EOS_TOKEN
def formatting_prompts_func(examples):
    texts = []
    # Iterate directly through the relevant columns from the CSV
    for target, input_text, stance in zip(examples["target"], examples["text"], examples["stance"]):
        # Create the instruction dynamically
        instruction = "Analyze the stance and identify the target in the following text."
        # Combine target and stance for the response
        output_combined = f"Target: {target}, Stance: {stance}"

        # Must add EOS_TOKEN, otherwise your generation will go on forever!
        text = alpaca_prompt.format(instruction, input_text, output_combined) + EOS_TOKEN
        texts.append(text)
    return { "text" : texts, }

from datasets import load_dataset
# Load the local CSV dataset
dataset = load_dataset("csv", data_files="train_dataset.csv", split="train")
dataset = dataset.map(formatting_prompts_func, batched = True,)
```

<a name="Train"></a>
### Train the model
Now let's use Huggingface TRL's `SFTTrainer`! More docs here: [TRL SFT docs](https://huggingface.co/docs/trl/sft_trainer). We do 60 steps to speed things up, but you can set `num_train_epochs=1` for a full run, and turn off `max_steps=None`. We also support TRL's `DPOTrainer`!

```python
from trl import SFTTrainer
from transformers import TrainingArguments
from unsloth import is_bfloat16_supported

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    dataset_num_proc = 2,
    packing = False, # Can make training 5x faster for short sequences.
    args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        # num_train_epochs = 1, # Set this for 1 full training run.
        max_steps = 60,
        learning_rate = 2e-4,
        fp16 = not is_bfloat16_supported(),
        bf16 = is_bfloat16_supported(),
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = "outputs",
        report_to = "none", # Use this for WandB etc
    ),
)
```

```python
# @title Show current memory stats
gpu_stats = torch.cuda.get_device_properties(0)
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
max_memory = round(gpu_stats.total_memory / 1024 / 1024 / 1024, 3)
print(f"GPU = {gpu_stats.name}. Max memory = {max_memory} GB.")
print(f"{start_gpu_memory} GB of memory reserved.")
```

```python
trainer_stats = trainer.train()
```

<a name="Inference"></a>
### Inference
Let's run the model! You can change the instruction and input - leave the output blank!

**[NEW] Try 2x faster inference in a free Colab for Llama-3.1 8b Instruct [here](https://colab.research.google.com/github/unslothai/notebooks/blob/main/nb/Unsloth_Studio.ipynb)**

```python
import pandas as pd
import re

# Load test dataset
test_dataset = pd.read_csv("test_dataset.csv")

# Function to parse model output to extract target and stance
def parse_model_output(output):
    try:
        # Expected format: "Target: {target}, Stance: {stance}"
        target_match = re.search(r"Target:\s*(.*?),\s*Stance:", output)
        stance_match = re.search(r"Stance:\s*(.*?)(?:\s*$|\n)", output)

        target = target_match.group(1).strip() if target_match else "Not Found"
        stance = stance_match.group(1).strip() if stance_match else "Not Found"

        return target, stance
    except Exception as e:
        print(f"Error parsing output: {output}")
        print(f"Error: {e}")
        return "Error", "Error"

# Prepare results storage
results = []

# Enable inference mode
FastLanguageModel.for_inference(model)

# Process each test example
for idx, row in test_dataset.iterrows():
    input_text = row['text']

    # Format input using the same template as training
    prompt = alpaca_prompt.format(
        "Analyze the stance and identify the target in the following text.",  # instruction
        input_text,  # input
        ""  # output - leave blank for generation
    )

    # Tokenize and generate
    inputs = tokenizer([prompt], return_tensors="pt").to("cuda")
    outputs = model.generate(
        **inputs,
        max_new_tokens=128,
        pad_token_id=tokenizer.eos_token_id,
        use_cache=True
    )
    # Decode output
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Extract the generated response part (after "### Response:")
    response_part = generated_text.split("### Response:")[-1].strip()

    # Parse the response to get predicted target and stance
    pred_target, pred_stance = parse_model_output(response_part)

    # Store results
    results.append({
        "Original Text": input_text,
        "Predicted Target": pred_target,
        "Predicted Stance": pred_stance,
        "Ground Truth Target": row['target'],
        "Ground Truth Stance": row['stance']
    })

    # Print progress
    if (idx + 1) % 10 == 0:
        print(f"Processed {idx + 1}/{len(test_dataset)} examples")
# Convert results to DataFrame and save to Excel
results_df = pd.DataFrame(results)
results_df.to_excel("test_results.xlsx", index=False)
print("\nResults saved to test_results.xlsx")

# Print some statistics
total = len(results_df)
target_matches = (results_df["Predicted Target"] == results_df["Ground Truth Target"]).sum()
stance_matches = (results_df["Predicted Stance"] == results_df["Ground Truth Stance"]).sum()
print(f"\nAccuracy Statistics:")
print(f"Target Accuracy: {target_matches/total*100:.2f}%")
print(f"Stance Accuracy: {stance_matches/total*100:.2f}%")
```

```python

import pandas as pd
import re

def evaluate_model(model, tokenizer, test_dataset, model_name=""):
    """
    Evaluate a model on the test dataset and return results
    """
    # Function to parse model output to extract target and stance
    def parse_model_output(output):
        try:
            # Expected format: "Target: {target}, Stance: {stance}"
            target_match = re.search(r"Target:\s*(.*?),\s*Stance:", output)
            stance_match = re.search(r"Stance:\s*(.*?)(?:\s*$|\n)", output)

            target = target_match.group(1).strip() if target_match else "Not Found"
            stance = stance_match.group(1).strip() if stance_match else "Not Found"

            return target, stance
        except Exception as e:
            print(f"Error parsing output: {output}")
            print(f"Error: {e}")
            return "Not Found", "Not Found"

    # Prepare results storage
    results = []

    # Enable inference mode
    FastLanguageModel.for_inference(model)

    print(f"\nEvaluating {model_name}...")

    # System prompt to guide the model's response format
    system_prompt = """You are a stance detection model. For the given text, you must:
1. Identify the target of the stance
2. Determine the stance (FAVOR, AGAINST, or NONE)
3. Format your response exactly as: "Target: <target>, Stance: <stance>"
For example:
Input: "I love using electric cars, they're great for the environment"
Response: "Target: electric cars, Stance: FAVOR"
"""

    # Process each test example
    for idx, row in test_dataset.iterrows():
        input_text = row['text']

        # Format input using the same template as training but include system prompt for base model
        if model_name == "Base Model":
            prompt = f"{system_prompt}\n\n" + alpaca_prompt.format(
                "Analyze the stance and identify the target in the following text.",
                input_text,
                ""
            )
        else:
            prompt = alpaca_prompt.format(
                "Analyze the stance and identify the target in the following text.",
                input_text,
                ""
            )

        try:
            # Tokenize and generate
            inputs = tokenizer([prompt], return_tensors="pt").to("cuda")
            outputs = model.generate(
                **inputs,
                max_new_tokens=128,
                pad_token_id=tokenizer.eos_token_id,
                use_cache=True,
                temperature=0.1  # Lower temperature for more consistent outputs
            )

            # Decode output
            generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

            # Extract the generated response part (after "### Response:")
            response_parts = generated_text.split("### Response:")
            if len(response_parts) > 1:
                response_part = response_parts[-1].strip()
            else:
                response_part = generated_text.strip()

            # Parse the response to get predicted target and stance
            pred_target, pred_stance = parse_model_output(response_part)

        except Exception as e:
            print(f"Error processing example {idx}: {e}")
            pred_target, pred_stance = "Error", "Error"

        # Store results
        results.append({
            "Original Text": input_text,
            "Predicted Target": pred_target,
            "Predicted Stance": pred_stance,
            "Ground Truth Target": row['target'],
            "Ground Truth Stance": row['stance']
        })

        # Print progress and save intermediate results every 100 rows
        if (idx + 1) % 10 == 0:
            print(f"Processed {idx + 1}/{len(test_dataset)} examples")

        if (idx + 1) % 100 == 0:
            # Save intermediate results
            interim_df = pd.DataFrame(results)
            interim_filename = f"test_results_{model_name.lower().replace(' ', '_')}_rows_1_to_{idx+1}.xlsx"
            interim_df.to_excel(interim_filename, index=False)
            print(f"\nIntermediate results saved to {interim_filename}")

            # Calculate and show interim statistics
            total = len(interim_df)
            target_matches = (interim_df["Predicted Target"] == interim_df["Ground Truth Target"]).sum()
            stance_matches = (interim_df["Predicted Stance"] == interim_df["Ground Truth Stance"]).sum()

            print(f"\nInterim Statistics (First {idx+1} examples):")
            print(f"Target Accuracy: {(target_matches/total*100):.2f}%")
            print(f"Stance Accuracy: {(stance_matches/total*100):.2f}%")

    # Convert final results to DataFrame
    results_df = pd.DataFrame(results)

    # Calculate final statistics
    total = len(results_df)
    target_matches = (results_df["Predicted Target"] == results_df["Ground Truth Target"]).sum()
    stance_matches = (results_df["Predicted Stance"] == results_df["Ground Truth Stance"]).sum()

    target_accuracy = target_matches/total*100
    stance_accuracy = stance_matches/total*100

    print(f"\n{model_name} Final Accuracy Statistics:")
    print(f"Target Accuracy: {target_accuracy:.2f}%")
    print(f"Stance Accuracy: {stance_accuracy:.2f}%")

    return results_df, target_accuracy, stance_accuracy

# Load test dataset
test_dataset = pd.read_csv("test_dataset.csv")

# First, evaluate the base model
print("\nLoading base model...")
base_model, base_tokenizer = FastLanguageModel.from_pretrained(
    model_name="unsloth/Meta-Llama-3.1-8B",
    max_seq_length=max_seq_length,
    dtype=dtype,
    load_in_4bit=load_in_4bit,
)

base_results, base_target_acc, base_stance_acc = evaluate_model(
    base_model,
    base_tokenizer,
    test_dataset,
    "Base Model"
)

# Save base model results
base_results.to_excel("test_results_base_model.xlsx", index=False)
print("\nBase model results saved to test_results_base_model.xlsx")
```

### GGUF / llama.cpp Conversion
To save to `GGUF` / `llama.cpp`, we support it natively now! We clone `llama.cpp` and we default save it to `q8_0`. We allow all methods like `q4_k_m`. Use `save_pretrained_gguf` for local saving and `push_to_hub_gguf` for uploading to HF.

Some supported quant methods (full list on our [Wiki page](https://github.com/unslothai/unsloth/wiki#gguf-quantization-options)):
* `q8_0` - Fast conversion. High resource use, but generally acceptable.
* `q4_k_m` - Recommended. Uses Q6_K for half of the attention.wv and feed_forward.w2 tensors, else Q4_K.
* `q5_k_m` - Recommended. Uses Q6_K for half of the attention.wv and feed_forward.w2 tensors, else Q5_K.
